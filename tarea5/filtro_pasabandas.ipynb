# Archivo de sonido

El nocturno 13 de Chopin está grabado en el archivo nocturne.wav. Este nocturno se caracteriza porque la melodía se encuentra en frecuencias altas. Específicamente, después de la nota musical La de 880hz. Bajo este La, los acordes de acompañamiento interpretan el bajo.

```python
#Leyendo archivo WAV. El archivo tiene dos canales
signal, sampler, enc = wavread('nocturne.wav')
signal = signal.T
print signal
print sampler 
print enc
```

```python
#Graficando señal de cada canal en el tiempo
n = len(signal[0])
print n
dt = 1/float(sampler)
print dt
time = arange(0, len(signal[0])/float(sampler), dt)
plot(time[0:4000],signal[0][0:4000])
show()
plot(time[0:4000],signal[1][0:4000])
show()
```

```python
#Realizando transformada de Fourier para cada canal
freq = []
fft_x = []
for i in range(len(signal)):
    fft_x.append(fft(signal[i]) /n )# FFT normalizado
    freq.append(fftfreq(n, dt)) # Recuperamos las frecuencias
    plt.plot(freq[i],np.abs(fft_x[i]))
    show()
```

Las frecuencias de filtrado se determinaron en la separación entre la melodía y el bajo. La freuencia para eliminar el ruido producido por los armónicos de las notas se determinó en 5000 hz. 

```python
#Filtrando señal
freq_low = 880.0
freq_high = 5000
for i in range(len(fft_x)):
    fft_x[i][np.abs(freq[i]) > freq_high ] = 0
    fft_x[i][np.abs(freq[i]) < freq_low] = 0
plt.plot(freq[0],np.abs(fft_x[0]))
```

```python
#Realizando ransformada inversa
clean_f = []
for i in range(len(fft_x)):
    clean_f.append(ifft(fft_x[i])) 
    clean_f[i] = real(clean_f[i])
clean_f = array(clean_f).T
clean_f = clean_f
print clean_f
plot(time[0:4000],clean_f[0:4000])
```

```python
#Recreando archivo de sonido
clean_f = clean_f*1000000 #Amplificando señal
outfile = Sndfile('salida.wav', 'w', Format('wav'), 2, sampler )
outfile.write_frames(clean_f)
outfile.close()
```

```python
#comprobando archivo de sonido
signal, sampler, enc = wavread('salida.wav')
print signal
print sampler 
print enc
```

Al reproducir el archivo, se puede observar que la melodía es más clara. Sin embargo, aun persisten los bajos puesto que estos resuenan en armónicos que entran dentro de las frecuencias de la melodía.
