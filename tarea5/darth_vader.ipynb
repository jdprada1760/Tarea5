# Archivo de sonido

Los efectos que se incluiran para modificar la voz a la de Darth Vader son tres. Amplificación de las frecuencias bajas del sonido de la grabación para hacer la voz más fuerte, aumento del tiempo de muestreo para la transformada de Fourier con el objetivo de crear un delay en la voz y bajar por tanto la frecuencia de la misma, y se le añadirá distorsión aleatoria a las amplitudes de la grabación.

```python
#Leyendo archivo WAV. El archivo tiene dos canales
signal, sampler, enc = wavread('darthin.wav')
signal = signal.T
print signal
print sampler 
print enc
```

```python
#Graficando señal de cada canal en el tiempo
n = len(signal[0])
print n
dt = 1/float(sampler)
print dt
time = arange(0, len(signal[0])/float(sampler), dt)
plot(time[0:4000],signal[0][0:4000])
show()
plot(time[0:4000],signal[1][0:4000])
show()
```

```python
#Realizando transformada de Fourier para cada canal
freq = []
fft_x = []
delay = 1.4 #Factor de retraso del sonido para dar efecto a la grabación
for i in range(len(signal)):
    fft_x.append(fft(signal[i]) /n )# FFT normalizado
    freq.append(fftfreq(n, delay*dt)) # Recuperamos las frecuencias
    plt.plot(freq[i],np.abs(fft_x[i]))
    show()
```

```python
#Creando canales 
freq_dim = 200.0 #Corte de frecuencias bajas que serán amplificadas
amp = 6 #Factor de amplificación
fft_x[0][np.abs(freq[0]) < freq_dim] = amp*fft_x[0][np.abs(freq[0]) < freq_dim]
fft_x[1][np.abs(freq[1]) < freq_dim] = amp*fft_x[1][np.abs(freq[1]) < freq_dim]
plt.plot(freq[0],np.abs(fft_x[0]))
```

```python
#Generando ditorsión aleatoria
amp_dist = 2.0/100.0   #Amplitud de distorsión
ruido = 3000 #Frecuencia donde el ruido se quiere sea mayor
fft_x[0][np.abs(freq[0]) < ruido] = fft_x[0][np.abs(freq[0]) < ruido] + random.random()*(freq[0][np.abs(freq[0]) < ruido]/ruido)*amp_dist
fft_x[1][np.abs(freq[1]) < ruido] = fft_x[1][np.abs(freq[1]) < ruido] + random.random()*(freq[1][np.abs(freq[1]) < ruido]/ruido)*amp_dist
plt.plot(freq[0],np.abs(fft_x[0]))
```

A cada canal se le asignará una distorsión propia para generar más sensación de artificialidad. Además, la distorsión tiene mayor amplitud a mayores frecuencias.


```python
#Realizando ransformada inversa
clean_f = []
for i in range(len(fft_x)):
    clean_f.append(ifft(fft_x[i])) 
    clean_f[i] = real(clean_f[i])
clean_f = array(clean_f)
clean_f = clean_f.T
print clean_f
plot(time[0:4000],clean_f[0:4000])
```

```python
#Recreando archivo de sonido
clean_f = clean_f*1000000 #Amplificando señal
outfile = Sndfile('darthout.wav', 'w', Format('wav'), 2, sampler/delay ) #Acá es donde más se debe tener en cuenta el delay en el tiempo
outfile.write_frames(clean_f)
outfile.close()
```
