```python
#Interpolacion con polinomios de Lagrange
# x = conjunto de datos en x
# y = conjunto de datos en y
# x0 = punto donde se quiere interpolar (sin restriccion)
def lagrangeInterpol(x, y, x0):
    i = 0
    j = 0
    npuntos = len(x)
    sum = 0
    for i in range(npuntos):
        
        mult = 1
        #Genera el polinomio que pasa por todos los puntos
        for j in range(npuntos):
            if j != i:
                mult = mult*((x0 - x[j])/(x[i] - x[j]))
                
        sum = sum + mult*y[i]
        
    return sum  

# x = conjunto de datos en x (ordenados)
# y = conjunto de datos en y
# x0 = punto donde se quiere interpolar (debe estar en el dominio dado por x)
def linearInterpol(x, y, x0):
    index = where( x == x0)
    #verifica que el punto no se encuentre en los datos dados
    if len(index[0]) == 0:
        index = where( x < x0)
        #puntos que definen la recta
        xi = x[(index[0])[len(index[0])-1]]
        xf = x[(index[0])[len(index[0])-1]+1]
        yi = y[(index[0])[len(index[0])-1]]
        yf = y[(index[0])[len(index[0])-1]+1]
        #pendiente de la recta
        m = (yf-yi)/(xf-xi)
        #interpolacion
        y0 = yi + m*(x0-xi)
        return y0
    return y[index]
```
